```python
import sys
import pandas as pd
sys.path.append('../')
```

```python
# read clustering results
kmeans = pd.read_csv('../data/clustering_results/kmeans_2_results_hourly_categories_nooutliers.csv')
kmeans
```

```python
# read the labeling data
labeling_data = pd.read_pickle('../data/preprocessing_final/labeling_df_hourly_processed.pkl')
labeling_data
```

```python
# concatenate the date and hour columns into one datetime object column
labeling_data['date'] = pd.to_datetime(labeling_data['date']) + labeling_data['hour'] * pd.to_timedelta('1H')
labeling_data.drop(columns=['hour'], inplace=True)
labeling_data['id'] = labeling_data['id'].astype(str)
labeling_data['date'] = labeling_data['date'].astype(str)
labeling_data
```

```python
# merge the labeling features on the clustering results based on id and date and keep only the raws of the clustering results
kmeans_visualizations = kmeans.merge(labeling_data, on=['id', 'date'], how='left')
kmeans_visualizations
```

```python
# there is one duplicate value that contains in "mood" one NaN and one 'HAPPY' value, we will keep the 'HAPPY' value, which is the last one
kmeans_visualizations = kmeans_visualizations[~kmeans_visualizations.duplicated(subset=['id', 'date'], keep='last')].reset_index(drop=True)
kmeans_visualizations.to_pickle('../data/labeling_visualizations/kmeans_categories_final_labeling.pkl')
kmeans_visualizations
```
