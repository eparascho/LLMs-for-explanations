```python
import pandas as pd
```

```python
# read preprocessed training data
data = pd.read_pickle('../data/clustering_df.pkl')
column_names = data.columns.values.tolist()  # raw data column names
dates = data['date']  # store dates
column_names.remove('date')  # remove date
column_names.append('cluster')  # cluster
```

```python
# store hdbscan results with preprocessed training data
hdbscan = pd.read_csv('../data/clustering_results/hdbscan_results.csv')
hdbscan.drop(hdbscan.columns[0], axis=1, inplace=True)  # drop the first column
hdbscan.columns = column_names  # rename columns
hdbscan.rename(columns={'WORK/SCHOOL': 'WORK_SCHOOL'}, inplace=True)
hdbscan.insert(1, 'date', dates)
hdbscan
```

```python
# store kmeans results with preprocessed training data
kmeans = pd.read_csv('../data/clustering_results/kmeans_results.csv')
kmeans.drop(kmeans.columns[0], axis=1, inplace=True)  # drop the first column
kmeans.columns = column_names  # rename columns
kmeans.rename(columns={'WORK/SCHOOL': 'WORK_SCHOOL'}, inplace=True)
kmeans.insert(1, 'date', dates)
kmeans
```

```python
# read the labeling data
labeling_data = pd.read_pickle('../data/labeling_df.pkl')
labeling_data
```

```python
# concatenate the date and hour columns into one datetime object column
labeling_data['date'] = pd.to_datetime(labeling_data['date']) + labeling_data['hour'] * pd.to_timedelta('1H')
labeling_data.drop(columns=['hour'], inplace=True)
labeling_data['id'] = labeling_data['id'].astype(str)
labeling_data
```

```python
# merge the labeling features on the clustering results with the preprocessed training data based on id and date and keep only the raws of the clustering results with the preprocessed training data
labeling_visualizations = hdbscan.merge(labeling_data, on=['id', 'date'], how='left')
labeling_visualizations.to_pickle('../data/labeling_visualizations/hdbscan_labeling.pkl')
labeling_visualizations = kmeans.merge(labeling_data, on=['id', 'date'], how='left')
labeling_visualizations.to_pickle('../data/labeling_visualizations/kmeans_labeling.pkl')
```
