```python
import sys
import pandas as pd
sys.path.append('../')
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
```

```python
# read data
data = pd.read_csv('../../data/clustering_results/kmeans_2_results_hourly_categories_nooutliers3.csv')
```

```python
# prepare data for clustering (store and then remove id)
clusters = data['cluster']
data.drop(columns=['id', 'date', 'cluster'], inplace=True)
```

```python
# coloring data points based on cluster
color_map = {0: '#9999ff', 1: '#ffff33'}
colors = clusters.map(color_map)
```

```python
# visualize data through TSNE
projection = TSNE().fit_transform(data)
plt.scatter(*projection.T, alpha=0.1, s=3, c=colors)
plt.show()
```

```python
# cluster 0 --> purple
print("Cluster 0 has", len(data[clusters == 0]), "data points")
# cluster 1 --> yellow
print("Cluster 1 has", len(data[clusters == 1]), "data points")
```
