This notebook creates the plot that reveal the cluster membership for each user.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/clustering_results/kmeans_2_results_hourly_categories_nooutliers3.csv')
data
```

```python
# rename the user identifiers
ids, unique = pd.factorize(data['id'])
data['id'] = ids + 1
data
```

```python
pivot_df = data.groupby(['id', 'cluster']).size().unstack(fill_value=0)

# create the stacked bar chart
colors= ['#9999ff', '#ffff33']
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6), color=colors)
plt.xlabel('Users')
plt.ylabel('Membership in clusters')
plt.xticks(fontsize=8)  # Change 10 to the desired font size
plt.show()
```
