```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the data
data = pd.read_csv('responses_processed.csv')
```

```python
# replace with appropriate level
levels = {
    1.0: 'Very unclear',
    2.0: 'Unclear',
    3.0: 'Neutral',
    4.0: 'Clear',
    5.0: 'Very clear',
    np.nan: 'Not answered'}
data['instructions'] = data['instructions'].replace(levels)
```

```python
# find the count of a column values
category_order = ['Very unclear', 'Unclear', 'Neutral', 'Clear', 'Very clear', 'Not answered']
data['instructions'] = pd.Categorical(data['instructions'], categories=category_order, ordered=True)
instructions_plot = data['instructions'].value_counts().reindex(category_order)
instructions_plot
```

```python
# plot a barchart
plt.figure(figsize=(8, 6))
instructions_plot.plot(kind='bar', color='#99ccff')
plt.xlabel('Clarity of the instructions')
plt.ylabel('Participants per clarity level')
plt.show()
```
