```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the data
data = pd.read_csv('responses_AB_paired_final_processed.csv')
```

```python
# find the count of a coumn values
group_plot = data['group'].value_counts(ascending=True)
group_plot
```

```python
# plot a horizontal barchart
plt.figure(figsize=(8, 6))
group_plot.plot(kind='bar', color='#99ccff')
for index, value in enumerate(group_plot):
    plt.text(index, value, str(value), ha='center', va='bottom')
plt.xlabel('Group')
plt.ylabel('Participants per group')
plt.show()
```
